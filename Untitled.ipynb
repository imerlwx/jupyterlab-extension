```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('recent-grads.csv')
```

```python
df.head(5)
```

```python
grouped = df.groupby('Major_category')['Median'].mean()
```

```python
plt.figure(figsize=(10,6)) # This line is used to adjust the size of the figure
grouped.sort_values().plot(kind='barh', color='skyblue') # We use kind='barh' for a horizontal bar plot

plt.xlabel('Median Salary') # Labeling the x-axis
plt.ylabel('Major Categories') # Labeling the y-axis
plt.title('Median Salary across Major Categories') # Giving the plot a title

plt.show() # This line is used to display the plot
```

```python
df_engineering = df[df['Major_category'] == 'Engineering']
```

```python
grouped_engineering = df_engineering.groupby('Major')['Median'].mean()
```

```python
plt.figure(figsize=(10,6))
grouped_engineering.sort_values().plot(kind='barh', color='skyblue')

plt.xlabel('Median Salary')
plt.ylabel('Engineering Majors')
plt.title('Median Salary across Engineering Majors')

plt.show()
```

```python
df_grouped = df.groupby('Major_category').agg({'ShareWomen': 'mean', 'Median': 'mean'})
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_grouped['ShareWomen'], df_grouped['Median'], alpha=0.5)

# Adding labels for x and y axis
plt.xlabel('Proportion of Women')
plt.ylabel('Median Salary')

# Adding a title
plt.title('Relationship between Proportion of Women and Median Salary')

# Display the plot
plt.show()
```
