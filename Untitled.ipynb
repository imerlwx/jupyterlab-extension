```python
import pandas as pd
```

```python
data = pd.read_csv('recent-grads.csv')
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
```

```python
median_salaries = data.groupby('Major_category')['Median'].median()
```

```python
plt.figure(figsize=(10,6))
median_salaries.sort_values().plot(kind='barh', color='skyblue')
plt.xlabel('Median Salary')
plt.title('Median Salary by Major Category')
plt.show()
```
